```python
dist = [0.8, 0.11, 0.14, 0.17, 0.2, 0.23, 0.26, 0.29, 0.32, 0.35]
```

```python
case_fname = 'nc.case'

fname, ext = case_fname.split('.')
new_case_fname = fname + '_.' + ext
```

```python
if os.path.isfile(new_case_fname):
    os.unlink(new_case_fname)
```

```python
f_new = open(new_case_fname, 'a')
with open(case_fname, 'r') as f_exist:
    
    count = 1
    modify = False
    for line in f_exist:
        
        if 'Chord' in line:
            f_new.write(line)
            modify = True
            
        if not modify:
            if count == 11:
                count += 1
                continue
            else:
                f_new.write(line)
        else:
            f_new.write(f"{dist[count-1]} {next_point_to_probe[f'chord{count}']} {next_point_to_probe[f'beta{count}']}\n")
            count += 1
            if count > 10:
                modify = False
        
f_new.close()
```
